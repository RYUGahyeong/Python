- 이디야 매장 주요 분포 지역 (매장수가 많은 상위 5개 구이름, 매장 개수 출력)
- 구별 브랜드 각각의 매장 개수 조회 (구이름, 브랜드이름, 매장 개수 출력)
- 구별 브랜드 각각의 매장 개수 조회 (구이름, 스타벅스 매장 개수, 이디야 매장 개수 출력)

제출 7.
- 관련 코드 & 실행 결과 (ipynb)

```python
# 스타벅스 매장 주요 분포지역
sql = """
    SELECT s.gu_name, COUNT(s.brand) AS brand_count
    FROM COFFEE_BRAND b
    INNER JOIN COFFEE_STORE s ON b.id = s.brand
    WHERE b.name = 'STARBUCKS'
    GROUP BY s.gu_name
    ORDER BY brand_count DESC
    LIMIT 5
"""
cur.execute(sql)
results = cur.fetchall()

for row in results:
    print(row)
```

```python
# 스타벅스 매장 주요 분포지역
sql = """
    SELECT s.gu_name, COUNT(s.brand) AS brand_count
    FROM COFFEE_BRAND b
    INNER JOIN COFFEE_STORE s ON b.id = s.brand
    WHERE b.name = 'EDIYA'
    GROUP BY s.gu_name
    ORDER BY brand_count DESC
    LIMIT 5
"""
cur.execute(sql)
results = cur.fetchall()

for row in results:
    print(row)
```

```python
# 구별 브랜드 각각의 매장 개수 조회 (구이름, 매장, 개수)
sql = """
    select s.gu_name, b.name, count(s.brand) 
    from COFFEE_BRAND  b, COFFEE_STORE s
    where b.id = s.brand
    group by s.gu_name, s.brand
    order by s.gu_name, b.name;
"""
cur.execute(sql)
results = cur.fetchall()

for row in results:
    print(row)
```

```python
# 구별 브랜드 각각의 매장 개수 조회 (구이름, 스타벅스 매장 개수, 이디야 매장 개수)
sql = """
    select s.gu_name, s.count STARBUCKS, e.count EDIYA 
    from (select gu_name, count(brand) count from COFFEE_STORE WHERE brand = 1 GROUP BY gu_name) s, 
        (select gu_name, count(brand) count from COFFEE_STORE WHERE brand = 2 GROUP BY gu_name) e 
    where s.gu_name = e.gu_name;
"""
cur.execute(sql)
results = cur.fetchall()

for row in results:
    print(row)
```

문제 8.

시각화 프로젝트를 위하여 다음의 규칙으로 쿼리하여 CSV 파일로 저장합니다. (Python 코드로 작업)

- 전체 데이터를 가져오는데, 각 스타벅스 매장별로 이디야 전체 매장정보가 매칭되어 있어야 합니다. (정렬 : s_id, e_id 순)
- 다음의 형식으로 저장되어야 합니다. (브랜드 이름, 칼럼 명 주의)
- 데이터 프레임 출력을 해주세요. 데이터 프레임 미출력시 감점입니다.

```python
sql = """
  select * 
  from (select id, 'STARBUCKS', name, gu_name, address, lat, lng
  from COFFEE_STORE WHERE brand = 1) s, 
  (select id, 'EDIYA',name, gu_name, address, lat, lng 
  from COFFEE_STORE WHERE brand = 2 GROUP BY gu_name) e
  order by s.id, e.id;
"""
cur.execute(sql)
result = cur.fetchall()
```

```python
df = pd.DataFrame(result)
df.columns = ["s_id","s_brand","s_name","s_gu", "s_address","s_lat","s_lng","e_id","e_brand","e_name","e_gu", "e_address","e_lat","e_lng"]
df.head(5)
```
